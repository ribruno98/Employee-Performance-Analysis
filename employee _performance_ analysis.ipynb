# Employee Performance Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

employee = pd.read_csv('combined_employee_dataset.csv')
#columns : age, years_experience, education_level, department, performance_score, performance_category

font1={'family':'serif', 'color':'blue', 'size':20}
font2={'family':'serif', 'color':'red', 'size':18}
font3={'family':'serif', 'color':'green', 'size':18}

sns.barplot(data=employee, x='performance_score', y='department', hue='department')
plt.title('Average performance by department', fontdict=font1)
plt.xlabel('Performance Score', fontdict=font3)
plt.ylabel('Departmet', fontdict=font2)
plt.show()

print(' ')
print(' ')
print(' ')
print(' ')

sns.barplot(data=employee, x='education_level', y='performance_score', hue='education_level')
plt.title("Average Performance by Education Level", fontdict=font1)
plt.xlabel('Education Level', fontdict=font3)
plt.ylabel('Performance Score', fontdict=font2)
plt.show()

print(' ')
print(' ')
print(' ')
print(' ')

sns.boxplot(data=employee, x='years_experience', y='performance_category', hue='performance_category')
plt.title("Experience Distribution by Performance Category", fontdict=font1)
plt.xlabel('Years Experience', fontdict=font3)
plt.ylabel('Performance Category', fontdict=font2)
plt.show()

print(' ')
print(' ')
print(' ')
print(' ')

sns.histplot(employee['age'], bins=30 ,kde=True)
plt.title('Employee Age Distribution', fontdict=font1)
plt.xlabel('Age', fontdict=font3)
plt.ylabel('Count', fontdict=font2)
plt.show()
```

Conclusions:



*   The department with the best performance score is Finance and the worst is the HR department.
*   The level of education that has the best performance score is Bachelor's level and the worst level is PhD.


*   There is not a big difference, whoever has more experience will have a better performance category, but from the graph you can see that the biggest category was Very good.
*   Most employees are between 40-50 years old.
